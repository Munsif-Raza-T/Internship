```python
# Importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
# Importing dataset
df = pd.read_csv('Iris.csv', index_col=0)
```

```python
# Looking for columns and rows
df.shape
```

This shows that this dataset has 150 records(rows) and 6 features(columns).

```python
# I/O data
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
# Creating dataframe to save data.
column_names = ['X_train', 'y_train', 'X_test', 'y_test']
train_test = pd.DataFrame(columns=column_names)
```

```python
# Using for loop creating 4 different subsets.
for i in np.arange(50,70,5):
    train_test.append(train_test_split(X, y, test_size=0.18, random_state=i))
```
