```python
import pickle
import matplotlib.pyplot as plt
import cv2
import face_recognition
```

```python
with open("Images", "rb") as fp:   # Unpickling
    images = pickle.load(fp)
```

```python
# Giving figure parameters.
plt.rcParams["figure.figsize"] = [50, 50]
plt.rcParams["figure.autolayout"] = True
```

```python
columns = 2 # Columns in figure
rows = 8 # rows in figure
index = 1 # index counter
for image in images:
    plt.subplot(rows, columns, index)
    plt.imshow(image)
    plt.axis('off')
    index = index+1
```

```python
image_encodings = []

for image in images:
    image_encodings.append(face_recognition.face_encodings(image))
```
