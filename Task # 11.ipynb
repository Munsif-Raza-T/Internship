```python
# Importing depandencies
from sklearn.datasets import make_biclusters
from sklearn.cluster import KMeans
```

```python
# Generating sample data
X, row, col = make_biclusters((100,3), 3)
```

```python
X
```

```python
# Using KMeans training the model for clustering.
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
```

```python
# Getting labels for each data point.
kmeans.labels_
```

```python
# Predicting a sample data points
kmeans.predict([[0, 0, 0], [55, 3, 15]])
```

```python
# Getting the centers of clusters.
kmeans.cluster_centers_
```
